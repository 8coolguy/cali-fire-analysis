```python
#trying to add population stats 
#trying to add income or prduction stats 
#trying to look at what type of economy they have by checking out farming data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl
```

```python
df =pd.read_csv(r'data/new_fire_weather_data.csv')
xl=pd.ExcelFile(r'data/cali_data_pile.xlsx')
pop_xl =pd.read_excel(r'data/population_by_county.xlsx')
```

```python

people_page =pd.read_excel(xl,xl.sheet_names[1])
```

```python
#add a columns into the df for population
#print(df.dtypes)
df.incident_dateonly_created=  pd.to_datetime(df.incident_dateonly_created)
pops =[]

for i in range(df.shape[0]):
    #print(df.incident_county.iloc[i])
    #print(df.incident_dateonly_created.iloc[i].year)
    
    total =0
    for j in range(len(pop_xl)): 

        
        if df.incident_county.iloc[i] ==pop_xl.county.iloc[j]:
            if df.incident_dateonly_created.iloc[i].year==pop_xl["year "].iloc[j] or (df.incident_dateonly_created.iloc[i].year >2018 and pop_xl["year "].iloc[j] ==2018):
                total+=int(pop_xl.population.iloc[j])
    pops.append(total)
    print(total)
pops 
```

```python

df["county_population"] =pops
df.county_population
```

```python
df.to_csv("data/new_fire_weather_data.csv")
```

```python
household_income_county=[]
my_dict =dict()
for i in range(len(df)):
    temp=df.incident_county.iloc[i]
    
    total =0
    if my_dict.__contains__(temp):
        total=my_dict.get(temp)
    else:  
        for j in range(len(people_page)):
            if temp == people_page.County.iloc[j]:
                my_dict[temp]=int(people_page["Median Household Income (2017)"].iloc[j])
                total=people_page["Median Household Income (2017)"].iloc[j]
    household_income_county.append(total)
                

    
    
```

```python
df["county_income"]=household_income_county
```

```python

df =df.drop(columns=["Unnamed: 0","Unnamed: 0.1"])
df.to_csv("data/new_fire_weather_data.csv")
```
